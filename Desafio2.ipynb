1. Considerando el experimento de seleccionar un país participante al azar, se pide
determinar las probabilidades de los siguientes sucesos:
a. pertenecer a África
b. pertenecer a Asia

d. pertenecer a Norteamérica

```python
import pandas as pd #FernandoSanchez

# Cargar el conjunto de datos FernandoSanchez
data = pd.read_csv('worldcup2014.csv')
total_teams = len(data)

# a. Pertenecer a África FernandoSanchez
africa_teams = len(data[data['continent'] == 'africa'])
prob_africa = africa_teams / total_teams

# b. Pertenecer a Asia FernandoSanchez
asia_teams = len(data[data['continent'] == 'asia'])
prob_asia = asia_teams / total_teams

# c. Pertenecer a Europa FernandoSanchez
europe_teams = len(data[data['continent'] == 'europe'])
prob_europe = europe_teams / total_teams

# d. Pertenecer a Norteamérica FernandoSanchez
north_america_teams = len(data[data['continent'] == 'northamerica'])
prob_north_america = north_america_teams / total_teams

# e. Pertenecer a Sudamérica FernandoSanchez
south_america_teams = len(data[data['continent'] == 'southamerica'])
prob_south_america = south_america_teams / total_teams

# f. Clasificar a segunda ronda FernandoSanchez
second_round_teams = len(data[data['clasificado'] == 1])
prob_second_round = second_round_teams / total_teams

# g. Ganar al menos un partido FernandoSanchez
juegos_ganados_teams = len(data[data['juegos_ganados'] >= 1])
prob_juegos_ganados_teams = juegos_ganados_teams / total_teams

# h. Pertenecer a África y clasificar a segunda ronda FernandoSanchez
h_teams = len(data[(data['continent'] == 'africa') & (data['clasificado'] == 1)])
prob_h = h_teams / total_teams

# i. Pertenecer a Europa y clasificar a segunda ronda FernandoSanchez
i_teams = len(data[(data['continent'] == 'europe') & (data['clasificado'] == 1)])
prob_i = i_teams / total_teams

# j. Pertenecer a Asia y clasificar a segunda ronda FernandoSanchez
j_teams = len(data[(data['continent'] == 'asia') & (data['clasificado'] == 1)])
prob_j = j_teams / total_teams

# k. Pertenecer a Sudamérica y clasificar a segunda ronda FernandoSanchez
k_teams = len(data[(data['continent'] == 'southamerica') & (data['clasificado'] == 1)])
prob_k = k_teams / total_teams

# l. Pertenecer a Norteamérica y clasificar a segunda ronda FernandoSanchez
l_teams = len(data[(data['continent'] == 'northamerica') & (data['clasificado'] == 1)])
prob_l = l_teams / total_teams

# m. Pertenecer a África o clasificar a segunda ronda FernandoSanchez
m_teams = len(data[(data['continent'] == 'africa') | (data['clasificado'] == 1)])
prob_m = m_teams / total_teams

# n. Pertenecer a Europa o clasificar a segunda ronda FernandoSanchez
n_teams = len(data[(data['continent'] == 'europe') | (data['clasificado'] == 1)])
prob_n = n_teams / total_teams

# o. Pertenecer a Asia o clasificar a segunda ronda FernandoSanchez
o_teams = len(data[(data['continent'] == 'asia') | (data['clasificado'] == 1)])
prob_o = o_teams / total_teams

# p. Pertenecer a Sudamérica o clasificar a segunda ronda FernandoSanchez
p_teams = len(data[(data['continent'] == 'southamerica') | (data['clasificado'] == 1)])
prob_p = p_teams / total_teams

# q. Pertenecer a Norteamérica o clasificar a segunda ronda FernandoSanchez
q_teams = len(data[(data['continent'] == 'northamerica') | (data['clasificado'] == 1)])
prob_q = q_teams / total_teams

# r. Pertenecer a África y haber ganado al menos un partido FernandoSanchez
r_teams = len(data[(data['continent'] == 'africa') & (data['juegos_ganados'] >= 1)])
prob_r = r_teams / total_teams

# s. Pertenecer a Europa y haber ganado al menos un partido FernandoSanchez
s_teams = len(data[(data['continent'] == 'europe') & (data['juegos_ganados'] >= 1)])
prob_s = s_teams / total_teams

# t. Pertenecer a Asia y haber ganado al menos un partido FernandoSanchez
t_teams = len(data[(data['continent'] == 'asia') & (data['juegos_ganados'] >= 1)])
prob_t = t_teams / total_teams

# u. Pertenecer a Sudamérica y haber ganado al menos un partido FernandoSanchez
u_teams = len(data[(data['continent'] == 'southamerica') & (data['juegos_ganados'] >= 1)])
prob_u = u_teams / total_teams

# v. Pertenecer a Norteamérica y haber ganado al menos un partido FernandoSanchez
v_teams = len(data[(data['continent'] == 'northamerica') & (data['juegos_ganados'] >= 1)])
prob_v = v_teams / total_teams

# w. Haber ganado al menos un partido y clasificar a segunda ronda FernandoSanchez
w_teams = len(data[(data['juegos_ganados'] >= 1) & (data['clasificado'] == 1)])
prob_w = w_teams / total_teams

#FernandoSanchez
'''Al imprimir los resultados nos damos cuenta de que la mayor cantidad de equipos pertenece a Europa, que la mitad de todos los equipos paso a segunda ronda.
Al igual notamos que el 71% gano al menos un partido.
La mayor probabilidad de pasar a segunda ronda las tiene europa y sudamerica quedando con la segunda posicion y aunque tiene menos de la mitad de los equipos que participan de europa.
Asia da la probabilidad del 0.0 para pasar a segunda ronda
Las probabiliades que incluyen la condicion o nos indica por individual que porcentaje pertenece al pais o paso a la siguiente ronda esto incluyendo al pais indicado y la mayor probabilidad es de 
europa esto nos afirma que influye la cantidad de equipos por pais que ingresan al torneo.
Tambien vemos que Europa tiene la mayor probabilidad de ganar al menos un partido.
Por ultimo vemos que los equipos que ganan al menos un partido y pasan a la siguiente ronda es del 50%'''
print("Probabilidad de pertenecer a Africa:", prob_africa)
print("Probabilidad de pertenecer a Asia:", prob_asia)
print("Probabilidad de pertenecer a Europa:", prob_europe)
print("Probabilidad de pertenecer a Norteamérica:", prob_north_america)
print("Probabilidad de pertenecer a Sudamérica:", prob_south_america)
print("Probabilidad de clasificar a segunda ronda:", prob_second_round)
print("Probabilidad de ganar al menos un partido:", prob_juegos_ganados_teams)
print("Probabilidad de pertenecer a Africa y clasificar a segunda ronda:", prob_h)
print("Probabilidad de pertenecer a Europa y clasificar a segunda ronda:", prob_i)
print("Probabilidad de pertenecer a Asia y clasificar a segunda ronda:", prob_j)
print("Probabilidad de pertenecer a Sudamérica y clasificar a segunda ronda:", prob_k)
print("Probabilidad de pertenecer a Norteamérica y clasificar a segunda ronda:", prob_l)
print("Probabilidad de pertenecer a Africa o clasificar a segunda ronda:", prob_m)
print("Probabilidad de pertenecer a Europa o clasificar a segunda ronda:", prob_n)
print("Probabilidad de pertenecer a Asia o clasificar a segunda ronda:", prob_o)
print("Probabilidad de pertenecer a Sudamérica o clasificar a segunda ronda:", prob_p)
print("Probabilidad de pertenecer a Norteamérica o clasificar a segunda ronda:", prob_q)
print("Probabilidad de pertenecer a Africa y haber ganado al menos un partido:", prob_r)
print("Probabilidad de pertenecer a Europa y haber ganado al menos un partido:", prob_s)
print("Probabilidad de pertenecer a Asia y haber ganado al menos un partido:", prob_t)
print("Probabilidad de pertenecer a Sudamérica y haber ganado al menos un partido:", prob_u)
print("Probabilidad de pertenecer a Norteamérica y haber ganado al menos un partido:", prob_v)
print("Probabilidad de haber ganado al menos un partido y clasificar a segunda ronda:", prob_w)


```
